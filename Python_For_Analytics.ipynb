```python
# case_df.loc[case_df['type'] == 'เสียง', ['district']].value_counts()
case_df.loc[case_df['type'] == 'เสียง']['district'].value_counts()
```

```python
case_df.loc[case_df['province'].fillna('').str.contains('กรุงเทพ|ฺBangkok')]['province'].value_counts()
```

```python
def map_bkk(x):
    if 'กรุงเทพ' in x:
        return 'กรุงเทพมหานคร'

case_df['province'] = case_df['province'].fillna('').apply(map_bkk)

case_df['province'].value_counts()
```

```python
bkk_df = case_df.loc[case_df['province'].fillna('').str.contains('กรุงเทพ|ฺBangkok')].reset_index()
bkk_df
```

```python
print(bkk_df.info())
bkk_df['star'].fillna(-1, inplace=True)
print(bkk_df.info())
bkk_df['star'] = bkk_df['star'].astype(int)
print(bkk_df.info())
```
